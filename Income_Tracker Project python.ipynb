```python
import json # should import json for the file

transactions = [] # the list of transaction

def load_userinfo(): # this will load a json file
    try:
        with open("transactions.json", "r") as file:
            return json.load(file)
    except FileNotFoundError: # will handle with error in case the file not found
        return []

def save_userinfo(): # this will save a json file
    with open("transactions.json", "w") as file:
        json.dump(transactions, file) # du

def add_transaction(): # this fun will take input and add trans
    amount = float(input("Enter the amount: "))
    category = input("Enter the category: \n - Food \n - Transportation \n - Entertainment \n - Salary \n  ")
    date = input("Enter the date: ")

    transaction = {"amount": amount, "category": category, "date": date} # will save data in dec
    transactions.append(transaction) # to add the data by using append
    save_userinfo()
    print("Transaction added successfully.")

def categorize_transaction(): # this fun to catg trans in a list of trans
    if not transactions:
        print("No transactions available ")
        return

    print("Select the transaction to categorize:")
    for idx, transaction in enumerate(transactions): # use enumetare fun to allow to iterate seq and track index
        print(f"{idx + 1}. Amount: {transaction['amount']} | Category: {transaction['category']} | Date: {transaction['date']}")

    try:
        index = int(input("Enter the transaction index: ")) - 1
        if 0 <= index < len(transactions): # len fun will return length
            category = input("Enter the new category: ")
            transactions[index]['category'] = category
            save_userinfo() # to update the trans data
            print("Transaction categorized successfully")

        else:
            print("Invalid transaction index")
    except ValueError:
        print("Invalid input! Please enter valid transaction index")

def view_transactions(): # this fun can make user to view their trans
    if not transactions: # if there no transcation will print there is no trans available
        print("No transactions available.")
        return

        print("Select the sorting option:") # the feature of view can sort trans
    print("1. Sort by Date") # by date
    print("2. Sort by Category") # by catag

    choice = input("Enter your choice: ") # if the user select

    if choice == "1": # will sort by date
        sorted_transactions = sorted(transactions, key=lambda t: t['date'])
        print("Transactions sorted by Date:")

    elif choice == "2": # will sort by catag
        sorted_transactions = sorted(transactions, key=lambda t: t['category'])
        print("Transactions sorted by Category:")
    else:
        print("Invalid choice! ") # if was not 1 or 2
        return
    for idx, transaction in enumerate(sorted_transactions): # sort list of dicts and print trans with index, amount, catg and date
    # enumerate fun generate index for item in the list
    # the transaction var used to access contents of the current dic
        print(f"{idx + 1}. Amount: {transaction['amount']} | Category: {transaction['category']} | Date: {transaction['date']}")

    print("Select an option:") # can update or delete the date of trans
    print("1. Update Transaction")
    print("2. Delete Transaction")
    sub_choice = input("Enter your choice: ")

def generate_reports(): #this fun will generate report
    if not transactions:
        print("No transactions available.") # if no data
        return

# report will calc total expenses, income, spend by catg and average

# the code below about how will calc
    total_expenses = 0
    total_income = 0
    categories = {}

    for transaction in transactions:
        amount = transaction["amount"]
        if amount < 0:
            total_expenses += amount # total of expenses
        else:
            total_income += amount # total of income

        category = transaction["category"]
        if category in categories:
            categories[category].append(amount)
        else:
            categories[category] = [amount]

    print("******* Reports *******")
    print(f"Total Expenses: {total_expenses}")
    print(f"Total Income: {total_income}")
    print()

    print("Spending by Category:")
    for category, amounts in categories.items():
        total = sum(amounts)
        count = len(amounts)
        average = total / count # calc the avg
        print(f"{category}: Total={total}, Count={count}, Average={average}")
    print()

# if user select exit will stop the program
def exit_program():
    save_userinfo()
    print("Thank you for using the Expense and Income Tracker!")
    quit() # this fun will exit the program

# load existing trans
transactions = load_userinfo()

# main loop
while True: # list of options what user wants prosess
    print("******* Expense and Income Tracker *******")
    print("1- Add Transaction")
    print("2- Categorize Transaction")
    print("3- View Transactions")
    print("4- Generate Reports")
    print("0- Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        add_transaction() # to add data
    elif choice == "2":
        categorize_transaction() # to catag trans
    elif choice == "3":
        view_transactions() # to view, edit, sort trans
    elif choice == "4":
        generate_reports() # to generate report
    elif choice == "0":
      exit_program() # exit
      break

    else:
        print("Invalid choice please try again !") # if select ivalid options
```
